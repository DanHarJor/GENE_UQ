```python
import numpy as np
import matplotlib.pyplot as plt

def discontinuous_rastrigin(x, y):
    A = 10
    rastrigin = x**2 - A * np.cos(2 * np.pi * x) + A + y**2 - A * np.cos(2 * np.pi * y) + A
    discontinuity = np.sign(x) * 100 + np.sign(y) * 100
    return rastrigin + discontinuity

x = np.linspace(-2.12, 2.12, 1000)
y = np.linspace(-2.12, 2.12, 1000)
X, Y = np.meshgrid(x, y)
Z = discontinuous_rastrigin(X, Y)


plt.figure(figsize=(10, 8))
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Discontinuous Rastrigin Function in 2D')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

plt.figure()
plt.plot(x, discontinuous_rastrigin(x,1))
```

```python
import numpy as np

def piecewise_linear_2d(x, y):

    if x < 0 and y < 0:
        return 2 * x + 3 * y + 1
    elif x >= 0 and y < 0:
        return -x + 2 * y + 2
    elif x < 0 and y >= 0:
        return x - y + 3
    else:
        return 0.5 * x + y - 1

# Generate a grid of x and y values
x_values = np.linspace(-10, 10, 400)
y_values = np.linspace(-10, 10, 400)
X, Y = np.meshgrid(x_values, y_values)
Z = np.array([piecewise_linear_2d(x, y) for x, y in zip(np.ravel(X), np.ravel(Y))]).reshape(X.shape)

# Create the plot
plt.figure(figsize=(10, 6))
cp = plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar(cp)
plt.xlabel('x')
plt.ylabel('y')
plt.title('2D Piecewise Linear Function')
plt.show()

plt.figure()
plt.plot(x_values,[piecewise_linear_2d(xi ,-5) for xi in x_values])
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

def piecewise_linear_2d(x, y):
    if x < 0 and y < 0:
        return 2 * x + 3 * y + 1
    elif x >= 0 and y < 0:
        return -x + 2 * y + 2
    elif x < 0 and y >= 0:
        return x - y + 3
    else:
        return 0.5 * x + y - 1

# Generate a grid of x and y values
x_values = np.linspace(-10, 10, 400)
y_values = np.linspace(-10, 10, 400)
X, Y = np.meshgrid(x_values, y_values)
Z = np.array([piecewise_linear_2d(x, y) for x, y in zip(np.ravel(X), np.ravel(Y))]).reshape(X.shape)

# Create the 3D plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.set_title('3D Plot of the Piecewise Linear Function')

plt.show()
```
