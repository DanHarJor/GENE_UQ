# Sparse Space 
The plan is adaptively sample with sparse space. Each time we sample we take the interpolation model and do montecarlo UQ to produce the distribution and compute the KL-divergence with the last distribution. Then when it converges within some threshold ie kl-divergence is close to 0 ie the distribution doesn't change. Then we stop sampling and check if it is close to the true distribution.
